```python
folder_path = 'output_files'
filelist = os.listdir(folder_path)
dfs = {}
for i, filename in enumerate(filelist):
  globals()[f'df{i+1}'] = pd.read_csv(path.join(folder_path, filename))
  print(f"df{i+1} assigned with data from {filename}")
  dfs[f'df{i+1}'] = globals()[f'df{i+1}']
```
